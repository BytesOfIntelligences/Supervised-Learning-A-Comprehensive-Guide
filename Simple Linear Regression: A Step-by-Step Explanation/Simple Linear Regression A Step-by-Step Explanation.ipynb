```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the Random Samples Datasets for Complete the Simple Linear Regression Implementation
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
```

```python
# Calculate the Mean value for x and Y
mean_for_x = np.mean(X) 
mean_for_y = np.mean(y)
```

```python
print("Mean Value of X", mean_for_x)
print("Mean Value of y", mean_for_y)
```

```python
rise = np.sum((X - mean_for_x)* (y - mean_for_y))
run = np.sum((X - mean_for_x)**2)
```

```python
print("The Value of Rise: ", rise)
print("The Value of Run: ", run)
```

```python
# Calculate the Slope for the simple linear regression
slope  = rise/run
print("The Solpe Value is : ", slope)
```

```python
# Calculate the intercept Value
intercept = mean_for_y - (slope * mean_for_x)
print("The intercepted Value is : ", intercept)
```

```python
# Create the visualization
plt.figure(figsize=(4, 3))
plt.scatter(X, y, color = "red", label = "Data Observation")
plt.plot(X, slope*X + intercept, color = "blue", label = "Simple Linear Regression Line (SLR)")
plt.legend()
plt.xlabel("Independent Data Points")
plt.ylabel("Dependent Data Points")
plt.title("Simple Linear Regression (SLR)")
plt.show()
```
